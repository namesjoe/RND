```python
import numpy as np
```

```python
n=100
coef, intercept = 21,3
X = 2 *  np.random.rand(n,1)
y =  coef * X  + intercept +  2*np.random.randn(n,1)
X_b=np.c_[np.ones((n,1)), X]
print(f'function y = {coef}x + {intercept} + noise')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(X, y, 'b.')
#plt.axis([0,2,0,30])
plt.show()
```

```python
from sklearn.linear_model import SGDRegressor
sgd_reg = SGDRegressor(n_iter = 50, eta0 = 0.1)
sgd_reg.fit(X,y)
```

```python
sgd_reg.intercept_, sgd_reg.coef_
print (f'predicted func y = {round(sgd_reg.coef_[0],2)}x + {round(sgd_reg.intercept_[0],2)}')
plt.plot(X, y, 'b.')
plt.plot(X, sgd_reg.predict(X), 'rx')
plt.axis([0,2,0,30])
plt.show()
```
